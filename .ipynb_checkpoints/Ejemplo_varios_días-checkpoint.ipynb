```python
# importamos las librerías que vamos a necesitar
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import numpy as np

# nuestra librería
import ist_lib as ist

folder='dst_stations/2015-03/'
month='201503'

lat=0.
lon=0.
comp='---'

# observatorios usados para el índice Dst
obs=['her','hon','kak','sjg']
# días en calma para el mes de marzo de 2015
qs=['10','30','05','14','09']

# variable para guardar todos los datos (1440 minutos) de los cuatro observatorios,
# los cinco días en calma
dat = np.zeros([len(obs),len(qs),1440])

# leer todos los días en calma del mes para cada observatorio (sólo la componente H (c1))
for o in obs:
    for q in qs:
        filename=folder + o + '/' + o + month + q + 'qmin.min'
        df=ist.read_im(filename=filename, lat=lat, lon=lon, comp=comp)
        dat[obs.index(o)][qs.index(q)][:]=df.c1
```

# Analicemos los datos de algunos días en calma

```python
# primer observatorio HER, primer día 10 (10-03-2015)
plt.plot(dat[0][0][:])
```

```python
# primer observatorio HER, segundo día 30 (30-03-2015)
plt.plot(dat[0][1][:])
```

```python
# primer observatorio HER, tercer día 05 (05-03-2015)
plt.plot(dat[0][2][:])
```
