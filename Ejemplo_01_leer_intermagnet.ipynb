```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

filename='dst_stations/2015-03/her/her20150310qmin.min'
names=['datetime','doy','c1','c2','c3','c4']
colspecs=[(0,23), (25,27), (32,40), (42,50), (52,60), (62,70)]
df=pd.read_fwf(filename,header=22,names=names, colspecs=colspecs)

date=pd.to_datetime(df.datetime.apply(str), format = '%Y-%m-%d %H:%M:%S.%f')
```

```python
fig, ax = plt.subplots()
ax.plot(date,df.c3)

# use formatters to specify major and minor ticks
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))
ax.xaxis.set_minor_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))
_ = plt.xticks(rotation=90)  
```
